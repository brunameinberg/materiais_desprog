# Checkpoint 10

```
for (int i = 0; i < n; i++){ //loop externo

    for (int j = 0; j < n; j++){ //loop interno

    }

}
```

Quantidade de iterações do loop externo (x):

Contador começa de 0 e aumenta de um em um enquanto for menor que n.

Depois de x-1 iterações, a condição ainda vale \ 
0 + (x - 1) < n \
(x - 1) < n \
x < n + 1


x = O(n), ou seja, x <= c_x n


Quantidade de iterações do loop interno (y):

Contador começa em 0 e aumenta de 1 em 1 enquanto for menor que n.

Depois de y-1 iterações, a condição ainda vale \ 
0 + (y - 1) < n \
(y - 1) < n \
y < n + 1

y = O(n), ou seja, y <= c_y n


Valores de i ao longo do loop externo:
i = 0, 1, 2, ....., (x - 1)


Limitante para as iterações de todas as execuções do loop interno:
c_y (0 + 1 + 2 + ... + (x - 1))  \

Soma de PA
- primeiro elemento 0
- último elemento x-1
- número de elementos x


= c_y (0 + x - 1) (x / 2) \
= c_y (x - 1) (x / 2) \
= c_y (x² - x) / 2 


Como x <= c_x n, pelas regras de simplificação podemos concluir que a complexidade é O(n²)

Contador começa em 1 e multiplica por 2 enquanto for menor que n.

Depois de x-1 iterações, a condição ainda vale \ 
1 * 2^(x - 1) < n \
2^(x - 1) < n \
2^(x) / 2 < n \
2^(x) < 2n \



2^x = O(n), ou seja, 2^x <= c_x n


Quantidade de iterações do loop interno (y):

Contador começa em 0 e aumenta de 1 em 1 enquanto for menor que i.

Depois de y-1 iterações, a condição ainda vale \

(y - 1) < i \
y < i + 1

Valores de i ao longo do loop externo:
i = 1, 2, 4, ....., 2^(x - 1)


Limitante para as iterações de todas as execuções do loop interno:
c_y (1 + 2 + 4 + ... + 2^(x - 1))

Soma de PG
- primeiro elemento 1
- razão da PG é 2
- número de elementos x


= c_y 1 (2^x - 1) / (2 - 1) \
= c_y (2^x - 1) \


Como 2^x <= c_x n, pelas regras de simplificação podemos concluir que a complexidade é O(n)
